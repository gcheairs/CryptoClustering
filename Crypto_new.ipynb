```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load data into DataFrame and set index
crypto_data = pd.read_csv('Resources/crypto_market_data.csv')
crypto_data.set_index('coin_id', inplace=True)

# Display summary statistics
print(crypto_data.describe())

# Prepare the data using StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(crypto_data)
scaled_df = pd.DataFrame(scaled_data, columns=crypto_data.columns, index=crypto_data.index)

# Display the first five rows of the scaled DataFrame
print(scaled_df.head())

# Find the best value for k using the elbow method on the original scaled data
k_values = list(range(1, 12))
inertia_values = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_df)
    inertia_values.append(kmeans.inertia_)

# Plot the elbow curve
elbow_data = {'k_values': k_values, 'inertia_values': inertia_values}
elbow_df = pd.DataFrame(elbow_data)
plt.plot(elbow_df['k_values'], elbow_df['inertia_values'], marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method to Find Best k')
plt.show()

# Answer: The best value for k can be determined visually from the elbow method plot.

# Cluster cryptocurrencies with K-Means using the original scaled data
best_k_original = 3  # Replace with the optimal value obtained from the elbow method
kmeans_original = KMeans(n_clusters=best_k_original, random_state=42)
kmeans_original.fit(scaled_df)
crypto_data_with_clusters_original = crypto_data.copy()
crypto_data_with_clusters_original['cluster'] = kmeans_original.labels_

# Create a scatterplot
crypto_data_with_clusters_original.plot.scatter(x='price_change_percentage_24h', y='price_change_percentage_7d', c='cluster', colormap='viridis')
plt.title('Clustered Cryptocurrencies (Original Scaled Data)')
plt.show()

# Optimize clusters with Principal Component Analysis (PCA)
pca = PCA(n_components=3)
pca_data = pca.fit_transform(scaled_df)
explained_variance_ratio = sum(pca.explained_variance_ratio_)

# Create a new DataFrame with the PCA data
pca_df = pd.DataFrame(pca_data, columns=['PC1', 'PC2', 'PC3'], index=crypto_data.index)

# Display the first five rows of the PCA DataFrame
print(pca_df.head())

# Find the best value for k using the elbow method on the PCA data
k_values_pca = list(range(1, 12))
inertia_values_pca = []

for k in k_values_pca:
    kmeans_pca = KMeans(n_clusters=k, random_state=42)
    kmeans_pca.fit(pca_df)
    inertia_values_pca.append(kmeans_pca.inertia_)

# Plot the elbow curve for PCA
elbow_data_pca = {'k_values': k_values_pca, 'inertia_values': inertia_values_pca}
elbow_df_pca = pd.DataFrame(elbow_data_pca)
plt.plot(elbow_df_pca['k_values'], elbow_df_pca['inertia_values'], marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method to Find Best k (PCA)')
plt.show()

# Answer: The best value for k when using PCA data can be determined visually from the elbow method plot.
# It may or may not differ from the best k-value found using the original data.

# Cluster cryptocurrencies with K-Means using the PCA data
best_k_pca = 3  # Replace with the optimal value obtained from the elbow method on PCA data
kmeans_pca = KMeans(n_clusters=best_k_pca, random_state=42)
kmeans_pca.fit(pca_df)
crypto_data_with_clusters_pca = crypto_data.copy()
crypto_data_with_clusters_pca['cluster'] = kmeans_pca.labels_

# Create a scatterplot using PCA data
crypto_data_with_clusters_pca.plot.scatter(x='PC1', y='PC2', c='cluster', colormap='viridis')
plt.title('Clustered Cryptocurrencies (PCA Data)')
plt.show()

# Answer: Using fewer features (PC1 and PC2) can impact the clustering result, as seen in the scatterplot.

# Determine the weights of each feature on each principal component
weights_df = pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2', 'PC3'], index=crypto_data.columns)

# Display the weights DataFrame
print(weights_df)
```
